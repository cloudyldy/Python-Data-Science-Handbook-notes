### 4.3.1 调整图形：线条的颜色与风格 

```python
plt.plot(x, np.sin(x - 0), color='blue')        # 标准颜色名称 
plt.plot(x, np.sin(x - 1), color='g')           # 缩写颜色代码（rgbcmyk） 
plt.plot(x, np.sin(x - 2), color='0.75')        # 范围在0~1的灰度值 
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # 十六进制（RRGGBB，00~FF） 
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB元组，范围在0~1 
plt.plot(x, np.sin(x - 5), color='chartreuse'); # HTML颜色名称
```

```python
plt.plot(x, x + 4, linestyle='-')  # 实线        
plt.plot(x, x + 5, linestyle='--') # 虚线        
plt.plot(x, x + 6, linestyle='-.') # 点划线        
plt.plot(x, x + 7, linestyle=':');  # 实点线
```

## 4.5 可视化异常处理

### 4.5.1 基本误差线 

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
%config InlineBackend.figure_format = 'svg'#显示矢量图
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr = dy, fmt = '.k')
```

```python
plt.errorbar(x, y, yerr = dy, fmt = 'o', color = 'k',
            ecolor = 'lightgray', elinewidth = 3, capsize = 0)
```

### 4.5.2 连续误差 

```python
from sklearn.gaussian_process import GaussianProcess 
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

gp = GaussianProcess(corr='cubic', theta0=1e-2, thetaL=1e-4, thetaU=1E-1,
                     random_start=100)        
gp.fit(xdata[:, np.newaxis], ydata) 

xfit = np.linspace(0, 10, 1000)        
yfit, MSE = gp.predict(xfit[:, np.newaxis], eval_MSE=True)        
dyfit = 2 * np.sqrt(MSE)  # 2*sigma~95%置信区间
```

```python
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color = 'gray')

plt.fill_between(xfit, yfit - dyfit, yfit + dyfit, color = 'gray', alpha = 0.2)
plt.xlim(0, 10)
```

## 4.6 密度图与等高线图

```python
def f(x, y):            
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x) 
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors = 'black')
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')        
plt.colorbar()
```

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',                   
           cmap='RdGy')        
plt.colorbar()        
plt.axis(aspect='image')
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')        
plt.clabel(contours, inline=True, fontsize=8) 
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',                   
           cmap='RdGy', alpha=0.5)        
plt.colorbar();
```
